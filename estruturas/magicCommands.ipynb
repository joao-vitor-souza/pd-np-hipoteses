# Magic Commands

### Os magic commands (comandos mágicos) são atalhos para funções específicas dos notebooks do Jupyter. A sintaxe básica de um magic command começa com um % (comandos na linha) ou um %% (comandos na célula), seguido do nome do comando. Os mais úteis usados para a Análise e Ciência de Dados estão listados a seguir:

-------------------------------------------------------------------------------------------------------

### 1. %lsmagic
#### Lista todos os magic commands existentes.

```python
%lsmagic
```

### 2. %timeit
#### Calcula o tempo médio de execução de uma operação específica ou da célula como um todo.

```python
import numpy as np

# Diferença de tempo entre uma estrutura Numpy e uma estrutura nativa do Python. A diferença chega a ser de duas ordens de magnitude.
%timeit np.arange(1000)**2
%timeit l = [x**2 for x in list(range(1000))]
```

### 3. %matplotlib notebook


#### Cria plotagens interativas diretamente no corpo da célula.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook

plt.plot(np.random.randn(10).cumsum())
```

### 4. %matplotlib inline

#### Também integra a plotagem na célula, mas diferente do comando anterior a plotagem não é interativa.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.scatter(np.arange(30), np.random.normal(10, 2, size=30))
```

### 5. %env

#### Permite o gerenciamento de todas as variáveis de ambiente do notebook.

```python
%env
```

### 5. %run
#### Executa programas externos ao notebook, rodando-os na célula atual. O programa deve estar na mesma pasta do notebook.

```python
%run ola_mundo.ipynb
```

### 6. %pycat
#### Detalha todo o conteúdo de um arquivo. Desde metadados ao código fonte.

```python
%pycat ola_mundo.ipynb
```

### 7. %whos
#### Mostra todas as variáveis ativas no notebook, seu tipo, informações e o valor armazenado quando possível.

```python
a = 1
b = 'n1'
c = [True]
d = (1, {2: 'n2'})
%whos
```

### 8. %reset
#### Exclui todas as variáveis do notebook.

```python
%reset
```

### 9. %pinfo
#### Parecido com %whos, mostra informações de uma variável específica.

```python
%pinfo d
```

### 10. %autosave
#### Habilita o salvamento automático e define o tempo para isso.

```python
# Salvando o arquivo a cada 180 segundos.
%autosave 180
```
